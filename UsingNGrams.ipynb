```python
import pandas as pd
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
import string

# Download NLTK resources
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

# Initialize WordNet lemmatizer
lemmatizer = WordNetLemmatizer()
import pandas as pd

from google.colab import drive
drive.mount('/content/drive')
df = pd.read_csv('/content/drive/MyDrive/Copy of Copy of shareddocgog.csv')

# Define punctuation set
punctuation_set = set(string.punctuation)

# Iterate over each row in the dataset
for index, row in df.iterrows():
    hindi_idiom = row['hindi_idiom']
    english_idiom = row['english_idiom']

    # Tokenization
    hindi_tokens = word_tokenize(hindi_idiom)
    english_tokens = word_tokenize(english_idiom)

    # Stopword removal and lowering
    stop_words = set(stopwords.words('english'))
    filtered_hindi_tokens = [word.lower() for word in hindi_tokens if word.lower() not in stop_words and word.lower() not in punctuation_set]
    filtered_english_tokens = [word.lower() for word in english_tokens if word.lower() not in stop_words and word.lower() not in punctuation_set]

    # Lemmatization
    lemmatized_english_tokens = [lemmatizer.lemmatize(word) for word in filtered_english_tokens]

    print("Hindi idiom:", hindi_idiom)
    print("English idiom:", english_idiom)
    print("\nTokenization:")
    print("Hindi tokens:", hindi_tokens)
    print("English tokens:", english_tokens)
    print("\nStopword Removal and Lowering:")
    print("Filtered Hindi tokens:", filtered_hindi_tokens)
    print("Filtered English tokens:", filtered_english_tokens)
    print("\nLemmatization:")
    print("Lemmatized English tokens:", lemmatized_english_tokens)
    print("\n--------------------------------------------\n")
```

```python
import spacy
import pandas as pd

# Load spaCy models for English
nlp_en = spacy.load('en_core_web_sm')

# Iterate over each row in the dataset
for index, row in df.iterrows():
    english_idiom = row['english_idiom']

    # POS Tagging
    english_idiom_doc = nlp_en(english_idiom)
    english_pos_tags = [(token.text, token.pos_) for token in english_idiom_doc]

    print("\nEnglish idiom:", english_idiom)
    print("\nPart-of-Speech (POS) Tagging:")
    print("English POS tags:", english_pos_tags)
    print("\n--------------------------------------------\n")
```

```python
from nltk.tokenize import word_tokenize
# Function to generate n-grams from a given sentence
def generate_ngrams(sentence, n):
    tokens = word_tokenize(sentence)
    ngrams = []
    for i in range(len(tokens) - n + 1):
        ngram = ' '.join(tokens[i:i+n])
        ngrams.append(ngram)
    return ngrams
# Function to detect Hindi idioms from a given sentence
def detect_hindi_idioms(input_sentence, dataset, n):
    input_ngrams = generate_ngrams(input_sentence, n)
    matched_idioms = []
    max_matched_ngrams = 0
    for index, row in dataset.iterrows():
        hindi_idiom = row['hindi_idiom']
        hindi_ngrams = generate_ngrams(hindi_idiom, n)
        matched_ngrams = set(input_ngrams).intersection(hindi_ngrams)
        if len(matched_ngrams) > max_matched_ngrams:
            matched_idioms = [hindi_idiom]
            max_matched_ngrams = len(matched_ngrams)
        elif len(matched_ngrams) == max_matched_ngrams:
            matched_idioms.append(hindi_idiom)
    return matched_idioms

# Example usage:
#input_sentence = "मुंबई जैसे महंगे जगह पर कम पैसों में काम करके गुजरा करना, जैसे- यह ऊंट के मुंह में जीरा है।"
#input_sentence="जब दोस्तों ने मिलकर छुपा खेला खेला, तो उनके एक दोस्त ने कहा, 'चोर चोर मौसेरे भाई, इसे कोई और नहीं जाना चाहिए।"
input_sentence="जब मुश्किल समयों में हमें आशा की कमी महसूस होती है, तो याद रखना कि भगवान के घर में देर हो सकती है, पर अंधेर नहीं।"
#input_sentence="जब उसने अपने वादों को तोड़ दिया और मुख में राम बगल में छूरी कर ली, तो उसकी बेवजह की चालाकी सभी को हैरान कर दी।"

n = 3  # Adjust 'n' for different n-grams
matched_idioms = detect_hindi_idioms(input_sentence, df, n)
if matched_idioms:
    print("Matched Hindi Idioms:")
    for idiom in matched_idioms:
        print("-", idiom)
else:
    print("No Hindi Idiom found in the input sentence.")
```
